```python
# Simple Phishing Email Detection Program

# Get email details
sender = input("Enter Sender Email: ").lower()
subject = input("Enter Subject: ").lower()
body = input("Enter Email Body: ").lower()

score = 0

# Suspicious sender domains
suspicious_domains = ["@gmail-security.com", "@paypal-secure.com", "@bankverify.com"]

# Suspicious keywords
suspicious_subject = ["urgent", "verify", "account", "winner", "lottery", "free", "payment"]
suspicious_body = [
    "click here",
    "verify your account",
    "login immediately",
    "update your password",
    "bank account",
    "credit card",
    "limited time",
    "claim prize"
]

# Check sender
for domain in suspicious_domains:
    if domain in sender:
        score += 2
        print("⚠ Suspicious sender detected.")
        break

# Check subject
for word in suspicious_subject:
    if word in subject:
        score += 1

# Check body
for phrase in suspicious_body:
    if phrase in body:
        score += 1

# Display result
print("\nPhishing Score:", score)

if score >= 4:
    print("Result: High Risk - Phishing Email Detected!")
elif score >= 2:
    print("Result: Medium Risk - Possibly a Phishing Email.")
else:
    print("Result: Low Risk - Email appears Safe.")
```
