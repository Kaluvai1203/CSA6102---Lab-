```python
import hashlib

def calculate_hash(file_name):
    sha256 = hashlib.sha256()

    with open(file_name, "rb") as file:
        while True:
            data = file.read(4096)
            if not data:
                break
            sha256.update(data)

    return sha256.hexdigest()

file_name = "evidence.txt"

try:
    with open(file_name, "x") as file:
        file.write("This is the original digital evidence.")
except FileExistsError:
    pass

original_hash = calculate_hash(file_name)

print("Original SHA-256 Hash:")
print(original_hash)

print("\nVerify the evidence after handling...")
input("Press Enter to continue...")

current_hash = calculate_hash(file_name)

print("\nCurrent SHA-256 Hash:")
print(current_hash)

if original_hash == current_hash:
    print("\nEvidence Integrity Verified.")
    print("Chain of Custody Maintained.")
else:
    print("\nWarning! Evidence has been altered.")
    print("Chain of Custody Compromised.")
```
