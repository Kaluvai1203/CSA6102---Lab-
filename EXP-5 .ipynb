```python
# Digital Forensic Investigation Report Generator

# Get case details from the user
case_id = input("Enter Case ID: ")
investigator = input("Enter Investigator Name: ")
evidence = input("Enter Evidence Description: ")

# Generate the report
print("\n" + "=" * 60)
print("      DIGITAL FORENSIC INVESTIGATION REPORT")
print("=" * 60)

print("\nCase Information")
print("-" * 30)
print("Case ID      :", case_id)
print("Investigator :", investigator)
print("Evidence     :", evidence)

print("\n1. Identification")
print("-" * 30)
print("Identify the digital evidence, incident, and affected devices.")

print("\n2. Preservation")
print("-" * 30)
print("Secure and preserve evidence to prevent modification.")
print("Maintain chain of custody and create forensic copies.")

print("\n3. Collection")
print("-" * 30)
print("Collect digital evidence using approved forensic tools.")
print("Record acquisition details and evidence sources.")

print("\n4. Examination and Analysis")
print("-" * 30)
print("Analyze the collected evidence for relevant artifacts,")
print("logs, deleted files, metadata, and user activities.")

print("\n5. Reporting and Presentation")
print("-" * 30)
print("Prepare a clear forensic report with findings,")
print("methods used, evidence, and final conclusions.")

print("\nConclusion")
print("-" * 30)
print("The investigation followed the standard five-stage")
print("digital forensic investigation lifecycle.")

print("\n" + "=" * 60)
print("End of Report")
print("=" * 60)
```
