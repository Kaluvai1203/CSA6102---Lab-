```python
import os
import time
from datetime import datetime

def show_metadata(file_name):
    stats = os.stat(file_name)

    print("\nFile Metadata")
    print("-" * 40)
    print("Creation Time :", datetime.fromtimestamp(stats.st_ctime))
    print("Modified Time :", datetime.fromtimestamp(stats.st_mtime))
    print("Access Time   :", datetime.fromtimestamp(stats.st_atime))
    print("-" * 40)
file_name = "evidence.txt"

with open(file_name, "w") as file:
    file.write("Original Evidence\n")

print("Sample file created.")
print("\nOriginal Metadata:")
show_metadata(file_name)
print("\nWaiting for 5 seconds...")
time.sleep(5)

with open(file_name, "a") as file:
    file.write("New trace added by user.\n")

print("\nFile modified.")
print("\nUpdated Metadata:")
show_metadata(file_name)

print("\nDemonstration Complete.")
print("Locard's Exchange Principle:")
print("Modifying the file left digital traces in its metadata.")
```
